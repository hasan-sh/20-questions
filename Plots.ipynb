### number of triples considered of the chosen triples

```python
from util import helpers
import matplotlib.pyplot as plt
import numpy as np
from tabulate import tabulate
```

```python
def getStats(runs):
    stats = {}
    s = []
    for i, run in enumerate(runs):
        games = run['games'].values()
        for game in games:
            s.append([helpers.parseTriple([game.get('entity')], 'uri')[0], game.get('noAnswers'), game['questionsAsked']])
            # print(game)
            # stats[helpers.parseTriple([game.get('entity')], 'uri')[0]] = { 'noAnswers': game.get('noAnswers'), 'questionsAsked': game['questionsAsked']}
            # stats[helpers.parseTriple([game.get('entity')], 'uri')[0]] = [game.get('noAnswers'), game['questionsAsked']]
    # return stats
    return s
```

```python
runs = helpers.readPickleBack('tournament_runs_00.pkl')
print('Before Completing to KG: ', )
stats1 = getStats(runs)
for s in stats1:#.items():
    print(s)
```

```python
runs = helpers.readPickleBack('after_tournament_runs_00.pkl')
print('After Completing to KG: \n')
stats2 = getStats(runs)

for s in stats2:#.items():
    print(s)
```

```python

labels = ['', 'noAnswers', 'questionsAsked']
format = 'latex'
for s in [stats1, stats2]:
    print(tabulate(s, headers=labels, tablefmt=format), '\n')
```

```python
X = list(stats1.keys())
X_axis = np.arange(len(X))

fig, ax = plt.subplots()


for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
                ax.get_xticklabels() + ax.get_yticklabels()):
        item.set_fontsize(10)

fig.set_size_inches(10, 5)

questionsAsked1 = [k['questionsAsked'] for k in stats1.values()]
questionsAsked2 = [k['questionsAsked'] for k in stats2.values()]

ax.bar(X_axis - 0.2, questionsAsked1, 0.4, label = 'questions asked (before)')
ax.bar(X_axis + 0.2, questionsAsked2, 0.4, label = 'questions asked (after)')

ax.yaxis.label.set_fontsize(10)
ax.xaxis.label.set_fontsize(10)

plt.xticks(X_axis, X)
# plt.xlabel("Groups")
plt.ylabel("Number of Questions Asked")
plt.title("Number of Questions Asked for each Entity")
plt.legend()
plt.show()
fig.savefig('questionsAsked.png')
```
